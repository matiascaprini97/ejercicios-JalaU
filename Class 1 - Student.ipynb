**1.- Convertidor de dias a segundos**

Pide al usuario ingresar un número de días y conviértelo a segundos.

````python
x = input("Ingrese el número de días que desea.")
y = int(x)*24
r = y*3600
print("La cantidad de segundos totales es " + str(r))
````

**2.- Descuento por Compras**

Crea un programa que solicite al usuario el monto total de su compra. Aplica un descuento del 10% al total de la compra. Imprima el monto despues del descuento.

````python
x = input("Ingrese el monto total de su compra.")
m = int(x)*10
d = m  /100
r = int(x) - d
print("El monto final es " + str(r))
````

**3.- Conversión de Moneda**

Desarrolla un programa que convierta una cantidad de dinero de dólares a euros. Pide al usuario la cantidad en dólares y la tasa de cambio actual.

````python
d = input("Ingrese la cantidad de dólares que desea cambiar")
t = input("Ingrese la tasa de cambio actual de dólar a euro")
r = int(d) * float(t)
print("Tienes " + str(r) + " euros en total.")
````
