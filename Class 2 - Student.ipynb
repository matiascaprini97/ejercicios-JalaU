### Ejercicio 1

Crea un programa que clasifique a una persona en una de las siguientes categorías según su edad:

*   Niño (0-12 años)
*   Adolescente (13-17 años)
*   Adulto (18-64 años)
*   Adulto mayor (65 años en adelante)    

```python
age = int(input("Enter your age in years."));
if age >= 0 and age < 13:
  print("You were classified as a 'Child'");
if age > 12 and age < 18:
  print("You were classified as a 'Teen'");
if age > 17 and age < 65:
  print("You were classified as a 'Adult'");
if age >= 65 :
  print("You were classified as a 'Older adult'");   
```
